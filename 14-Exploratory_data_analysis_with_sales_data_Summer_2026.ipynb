### Question 1: Import Data and Store-wise Revenue Analysis (4 Marks)

Import `pandas` and `numpy`. Download the dataset `sales.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `sales.csv` into a DataFrame. Then, find the top 5 stores with the highest total sales and the bottom 5 stores with the lowest total sales.

**Tasks:**
- Aggregate total sales by store_id
- Rank stores by total revenue
- Compare average sales per department within those stores

```python
import pandas as pd
import numpy as np
from IPython.display import display
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("sales.csv")

# Obtaining total sales for each store
store_sales = df.groupby('store_id')['weekly_sales'].sum()
store_sales = store_sales.reset_index()
store_sales.columns = ['store_id', 'total_sales']

# Getting the top and bottom five stores after sorting
sorted_stores = store_sales.sort_values(by='total_sales',ascending=False)
top_5 = sorted_stores.head(5)
bottom_5 = sorted_stores.tail(5)
print("\nTop 5 stores")
display(top_5)
print("\nBottom 5 stores")
display(bottom_5)

# Getting top and bottom store IDs
top_store_ids = top_5['store_id']
bottom_store_ids = bottom_5['store_id']

# Getting original data for top and bottom stores
mask1 = df['store_id'].isin(top_store_ids)
top_store_data = df[mask1]
mask2 = df['store_id'].isin(bottom_store_ids)
bottom_store_data = df[mask2]

# Average department sales
top_dept_avg = (top_store_data.groupby('department')['weekly_sales'].mean().reset_index())
top_dept_avg.columns = ['department', 'average_sales']
bottom_dept_avg = (bottom_store_data.groupby('department')['weekly_sales'].mean().reset_index())
bottom_dept_avg.columns = ['department', 'average_sales']
print("\nAverage sales per department in top 5 stores")
display(top_dept_avg)
print("\nAverage sales per department in bottom 5 stores")
display(bottom_dept_avg)
```

### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly weekly_sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python
#Mean and standard deviation of weekly sales of each department
dept_stats = df.groupby('department')['weekly_sales'].agg(['mean', 'std'])
dept_stats = dept_stats.reset_index()
# Coefficient of Variation
dept_stats['cv'] = dept_stats['std'] / dept_stats['mean']
#Sorting the departments and getting most stable and most volatile
stable_departments = dept_stats.sort_values(by='cv')
top_stable = stable_departments.head(5)
volatile_departments = dept_stats.sort_values(by='cv', ascending=False)
top_volatile = volatile_departments.head(5)

print("\nMost Stable Departments")
display(top_stable)
print("\nMost Volatile Departments")
display(top_volatile)
```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
# Average weekly sales on holidyas and non-holidays
holiday_avg = df[df['is_holiday'] == True]['weekly_sales'].mean()
nonholiday_avg = df[df['is_holiday'] == False]['weekly_sales'].mean()

print("Average Holiday Sales:", holiday_avg)
print("Average Non-Holiday Sales:", nonholiday_avg)

#Percentage change
percentage_change = ((holiday_avg - nonholiday_avg) / nonholiday_avg) * 100
print("\nPercentage Change:", percentage_change, "%")

#Department-wise holiday and non-holiday sales
holiday_dept = (df[df['is_holiday'] == True].groupby('department')['weekly_sales'].mean())
nonholiday_dept = (df[df['is_holiday'] == False].groupby('department')['weekly_sales'].mean())

# Combine into one table
dept_comparison = pd.DataFrame({'holiday_avg': holiday_dept,'nonholiday_avg': nonholiday_dept}).reset_index()

# Percentage change by department
dept_comparison['percentage_change'] = ((dept_comparison['holiday_avg'] - dept_comparison['nonholiday_avg'])/ dept_comparison['nonholiday_avg']) * 100

# Departments benefiting most
top_departments = dept_comparison.sort_values(
    by='percentage_change',
    ascending=False
)

print("Departments benefiting most from holidays")
display(top_departments.head(5))
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
# Convert date column to datetime format
df["date"] = pd.to_datetime(df["date"])

# Extract month number
df["month"] = df["date"].dt.month

# Calculate average monthly sales
monthly_sales = df.groupby("month")["weekly_sales"].mean().reset_index()

# List of month names
months = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

# Convert month numbers to month names
monthly_sales["month"] = monthly_sales["month"].apply(lambda x: months[x - 1])

# Display average monthly sales
print("Average Monthly Sales:\n")
print(monthly_sales)

# Highest sales month
highest_month = monthly_sales.loc[monthly_sales["weekly_sales"].idxmax()]
print("\nHighest Sales Month:")
print(highest_month)

# Lowest sales month
lowest_month = monthly_sales.loc[monthly_sales["weekly_sales"].idxmin()]
print("\nLowest Sales Month:")
print(lowest_month)
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
#Group by store_id and department
store_dept_sales = df.groupby(["store_id", "department"])["weekly_sales"].sum().reset_index()

# Total revenue
total_revenue = store_dept_sales["weekly_sales"].sum()

# Percentage contribution
store_dept_sales["percentage_contribution"] = (store_dept_sales["weekly_sales"] / total_revenue) * 100

# Top-performing combination
top = store_dept_sales.loc[store_dept_sales["weekly_sales"].idxmax()]

# Worst-performing combination
worst = store_dept_sales.loc[store_dept_sales["weekly_sales"].idxmin()]

# Display results
print("Top-Performing Combination:")
print(f"Store: {top['store_id']}")
print(f"Department: {top['department']}")
print(f"Revenue: {top['weekly_sales']}")
print(f"Percentage Contribution: {top['percentage_contribution']:.2f}%")

print("\nWorst-Performing Combination:")
print(f"Store: {worst['store_id']}")
print(f"Department: {worst['department']}")
print(f"Revenue: {worst['weekly_sales']}")
print(f"Percentage Contribution: {worst['percentage_contribution']:.2f}%")
```

### Question 6: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
from sklearn.datasets import load_digits

# Load the MNIST (Digits) dataset
digits = load_digits()

# Separate features and target labels
X = digits.data
y = digits.target

# Print the shapes
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`).
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

# Initialize K-Means with 10 clusters
kmeans = KMeans(n_clusters=10, random_state=42)

# Fit the model on the feature data
kmeans.fit(X)

# Store the cluster labels
kmeans_labels = kmeans.labels_

# Display the fitted model
print("K-Means Model:")
print(kmeans)

# Display the cluster labels
print("\nCluster Labels:")
print(kmeans_labels)
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster.
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

# Create an array for mapped labels
mapped_labels = np.zeros_like(kmeans_labels)

# Map each cluster to the most frequent true label
for i in range(10):
    mask = (kmeans_labels == i)
    mapped_labels[mask] = mode(y[mask], keepdims=True).mode[0]

# Calculate the macro F1 score
f1 = f1_score(y, mapped_labels, average="macro")

# Print the F1 score
print("F1 Score:", f1)
```
